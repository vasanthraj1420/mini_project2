```python
import yaml
import pandas as pd
import os
all_data=[]
folder_path=r"C:\Users\vasan\Desktop\New folder"
for folder in os.listdir(folder_path):
    folder_n=os.path.join(folder_path,folder)
    if os.path.isdir(folder_n):
        for foldername in os.listdir(folder_n):
            if foldername.endswith(".yaml"):
                file_path=os.path.join(folder_n,foldername)
                try:
                    with open(file_path,"r")as yaml_file:
                        data=yaml.safe_load(yaml_file)
                    if isinstance(data,list):
                        all_data.extend(data)
                except Exception as e:
                    print(f"error{file_path}:{e}")
```

```python
df=pd.DataFrame(all_data)
df
```

```python
path=(r"C:\Users\vasan\Desktop\Stock_ Analysis_miniproject_2\all_datas.csv")
df.to_csv(path,index=False)
```

```python
Ticker=pd.read_csv(r"C:\Users\vasan\Desktop\New folder\Ticker.csv")
Tic=Ticker['Ticker']
for i in Tic:
    rows = df[df['Ticker'] == i]
    path=(f"C:/Users/vasan/Desktop/project 2/{i}.csv")
    rows.to_csv(path,index=False)
```
